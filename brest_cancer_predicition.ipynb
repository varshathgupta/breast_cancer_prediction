```python
# M and B are changed to 1& 0 for analysis
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

```python
df['diagnosis'].unique()
```

```python
# Prdections Starts
```

```python
X = df.drop('diagnosis', axis=1)
X.head()
```

```python
y = df['diagnosis']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler 
ss =StandardScaler()
```

```python
X_train = ss.fit_transform(X_train) # It will make  data to study or fit
X_test = ss.transform(X_test) # It will test the data
```

```python
X_train
```

```python
X_test
```

```python
# ML Models
```

```python
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred= lr.predict(X_test)
y_pred
```

```python
y_test
```

```python
# To see accuarcy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
lr_acc = accuracy_score(y_test,y_pred)
lr_acc
```

```python
results = pd.DataFrame()
results
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression Method'],'Accuracy':[lr_acc]})
results = pd.concat([results,tempResults])
results = results[['Algorithm', 'Accuracy']]
results
```

```python
# Decision Tree Classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
dtc_acc =accuracy_score(y_test,y_pred)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision Tree Classifier'],'Accuracy':[dtc_acc]})
results = pd.concat([results,tempResults])
results = results[['Algorithm', 'Accuracy']]
results
```
